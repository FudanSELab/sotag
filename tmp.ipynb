```python
url='https://stackoverflow.com/tags/javascript/info'
html=requests.get(url,headers=user_agent).content
html
```

```python
soup = BeautifulSoup(html, 'html.parser', from_encoding='utf-8')
text_div = soup.find('div', class_="s-prose js-post-body") 
print(text_div.prettify()) #   页面中间一列的全部信息
```

```python
short_div = text_div.find('div', class_="welovestackoverflow")
print(short_div.prettify()) 
```

```python
p = short_div.find('p')
short_text = p.text
```

```python
info_html = ""
info_html += str(short_div)
long_text = ""
```

```python
for item_div in short_div.find_next_siblings():
    item_str = str(item_div)
    long_text += item_div.text + "\n"
    info_html += str(item_div)
```
